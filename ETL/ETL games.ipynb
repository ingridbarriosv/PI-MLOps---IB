# ETL Steam Games

```python
import pandas as pd
import json
import pyarrow
import ast
```

```python
#Se carga el dataset para crear el DataFrame que corresponde 
with open('DB Steam/steam_games.json', 'r') as file:
    data = [json.loads(line) for line in file]

dataGames = pd.json_normalize(data)
dfGames = pd.DataFrame(dataGames)
```

```python
dfGames
```

```python
#Se eliminan las filas que se encuentren vacias, y tambiÃ©n algunas columnas que no son necesarias para este caso
dfGames = dfGames.dropna(how = 'all')
dfGames.drop(['app_name','early_access','url','reviews_url','tags','publisher','specs'], axis=1, inplace = True)
dfGames.reindex()
```

```python
#Se extrae solamente el aÃ±o de la columna 'release_date' en una nueva columna cambiando el tipo de dato a int
dfGames['release_date'] = pd.to_datetime(dfGames['release_date'], format = 'mixed', errors = 'coerce')
dfGames['release_year'] = dfGames['release_date'].dt.year.fillna(0).astype(int)
```

```python
dfGames.head()
```

```python
#Se puede observar que en la columna price hay diferentes tipos de datos
dfGames['price'].apply(type).unique()
```

```python
#Se lista cuales son los valores tipo str de la columna price
dfGames['price'].loc[dfGames['price'].apply(type) == str].unique()
```

```python
'''Como queremos saber los precios por aÃ±o, entonces cambiamos los valores de la columna
price que nos indiquen que son gratis al valor de 0 y los que son pagos, a su valor correspondiente'''
dfGames.loc[dfGames['price'].str.contains('Free', case = False) & dfGames['price'].notna(), 'price'] = 0.0
dfGames.loc[dfGames['price'].str.contains('install', case = False) & dfGames['price'].notna(), 'price'] = 0.0
dfGames.loc[dfGames['price'].str.contains('play', case = False) & dfGames['price'].notna(), 'price'] = 0.0
dfGames.loc[dfGames['price'].str.contains('third', case = False) & dfGames['price'].notna(), 'price'] = 0.0
dfGames.loc[dfGames['price'].str.contains('449', case = False) & dfGames['price'].notna(), 'price'] = 449.0
dfGames.loc[dfGames['price'].str.contains('499', case = False) & dfGames['price'].notna(), 'price'] = 499.0
```

```python
dfGames['price'].apply(type).unique()
#Ya podemos ver que no hay valores de tipo str
```

```python
#Se cambia el nombre de la columna 'id' a 'item_id' para poder relacionar con los otros datasets
dfGames = dfGames.rename(columns = {'id': 'item_id'})
```

```python
dfGames
```

```python
#En este punto se transforma el dataframe a un formato parquet
dfGames.to_parquet('DB Steam/steam_games.parquet', index = False)
```

```python
dfGames.head()
```

```python
# Se crea un dataframe que contenga unicamente el Id de los juegos y tambiÃ©n su precio.
dfGamesPrice = dfGames[['item_id', 'price']]
dfGamesPrice.head()
```

```python
dfGamesPrice.head(20)
```

```python
# Podemos ver que hay valores del precio que son nulos, asÃ­ que se eliminarÃ¡n
dfGamesPrice = dfGamesPrice.dropna(subset = 'price')
dfGamesPrice.head(20)
```

```python
# Se transforma el dataframe creado a formato parquet
dfGamesPrice.to_parquet('DB Steam/GamesPrice.parquet', index = False)
```

```python
# Se crea un dataframe que contenga los generos de los juegos junto con su 'item_id'
dfGamesGenre = dfGames[['item_id', 'genres', 'release_year']]
dfGamesGenre.head()
```

```python
# Podemos ver que hay valores NaN en la columna 'genres', asÃ­ que se eliminan los juegos que no tengan especificado su genero
dfGamesGenre = dfGamesGenre.dropna(subset = 'genres')
```

```python
dfGamesGenre.head(20)
```

```python
# Se hace una comprensiÃ³n de lista que cambia cada elemento '&' por 'and' en la columna 'genres'
dfGamesGenre['genres'] = dfGamesGenre['genres'].apply(lambda x: [element.replace('&amp;', 'and') for element in x]) 
dfGamesGenre.head()
```

```python
# Seguidamente se transforma el dataframe a formato parquet 
dfGamesGenre.to_parquet('DB Steam/GamesGenre.parquet', index = False)
```

```python
# Se desanidan los datos de la columna 'genres' que estan anidados en listas con el mÃ©todo explode.
dfGamesGenre_sep = dfGamesGenre.explode('genres').reset_index(drop = True)
dfGamesGenre_sep.head()
```

```python
# Convertimos el dataframe con los generos desanidados a formato parquet
dfGamesGenre_sep.to_parquet('DB Steam/GamesGenre_sep.parquet', index = False)
```

```python
# Se crea un dataframe con la informaciÃ³n de los desarrolladores de los juegos.
# TambiÃ©n se eliminan los juegos que no tengan especificado su desarrollador.
dfGamesDevs = dfGames[['item_id', 'developer','release_year']]
dfGamesDevs = dfGamesDevs.dropna(subset = 'developer')
dfGamesDevs.head()
```

```python
# Lo convertimos a parquet como todos los anteriores 
dfGamesDevs.to_parquet('DB Steam/GamesDevs.parquet', index = False)
```
